```python
import matplotlib.pyplot as pl
```

```python
arr3 = ((np.arange(0,361))*(np.pi))/180
```

```python
arr4 = np.arange(0,361)
sin3 = np.sin(arr3)
cos3 = np.cos(arr3)
```

```python
pl.plot(arr4, sin3, '#0CF3B9', linewidth=2) #cyan
pl.plot(arr4, cos3,'#F30C46', lw=2) #pink
pl.plot(arr4, cos3-sin3, '#6900FF', lw=1, ls='dotted') #purple
pl.xlabel('Angle in degree')
pl.ylabel('Value')
pl.show()
```

```python
arr5 = np.linspace(1, 100, 20)
pl.plot(arr5, np.log(arr5), '#FFA000', lw=2, marker='D', markersize='3.7', markeredgecolor='#B48E4B') #orange
```

```python
#eg8.1
arr6 = np.arange(1, 21, 1.25)
log6 = np.log(arr6)
pl.plot(arr6, log6)
pl.xlabel('Random Values')
pl.ylabel('Logarithm Values')
pl.show
```

```python
#eg8.1 cond.
cos6 = np.cos(arr6)
pl.plot(arr6, log6, 'bd', ls='solid')
pl.plot(arr6, cos6, 'ko', ls='dashdot')
pl.show()
```
